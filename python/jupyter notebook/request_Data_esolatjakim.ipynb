```python
import requests
```

```python
urlStr = "https://www.e-solat.gov.my/index.php"
urlParams = {
    "r": "esolatApi/takwimsolat",
    "zone": "WLY01", ## WLY01 = WP KL & Putrajaya
    ## use period=duration to get specific timeframe within 1 year
    "period": "duration"
    ## use period=year or period=month or period=week for current year/month/week
    # "period": "year"
}
urlData = {
    # MUST DEFINE THESE if using period=duration in urlParams
    "datestart": "2018-01-01",
    "dateend": "2018-12-31"
}
```

```python
req = requests.post(url= urlStr, params=urlParams, data=urlData)
```

```python
resJson = req.json()
```

```python
eSolatJson = resJson["prayerTime"]
```
