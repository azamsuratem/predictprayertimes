```python
import pandas as pd
from tabula import read_pdf
```

```python
## Manual extract PDF Table for Jan 2024
## 1. Browse https://www.islamicfinder.org/prayer-times/
## 2. Under section "Prayer Times in <Location>", find at bottom left for "JAKIM Change", then click "Change"
## 3. Current location "Kuala Lumpur", Latitude and Longitude follow Muslim Pro i.e. Lat = 3.1499222, Long = 101.6944619
## 4. Click "Save & Update"
## 5. Scroll down to section "Prayer Timetable", dropdown select "January" and "2024", then "Monthly"/"Yearly" make sure "Monthly" is selected
## 6. Click "Download", which brings to print section
## 7. "Printer" select "Save as PDF" or "Microsoft Print to PDF", "Pages" select the input field and type "2" to select only Page 2
## 8. Click "Save", and give file name using this standard: "islamicfinder_YYYY_MM.pdf", also select folder "pdfdata"
## 9. Use the file path string from step 8 into variable pdfPathMMMYYYY i.e. pdfPathMMMYYYY = "pdfdata/islamicfinder_2024_01.pdf"
```

```python
pdfPathJan2024 = "pdfdata/islamicfinder_2024_01.pdf"
IslamicFinderPDFJan2024 = read_pdf(pdfPathJan2024, pages=1)
IslamicFinderDFJan2024 = IslamicFinderPDFJan2024[1]
IslamicFinderDFJan2024
```

```python
dfJan2024 = IslamicFinderDFJan2024[["January","Fajr","Dhuhr","Asr","Maghrib","Isha"]]
dfJan2024 = dfJan2024.drop(12)
dfJan2024.loc[:,"January"] = pd.to_datetime(dfJan2024["January"] + " Jan 2024", format="%d %b %Y").dt.date
dfJan2024.loc[:,"Fajr"] = pd.to_datetime(dfJan2024["Fajr"], format="%I:%M %p").dt.time
dfJan2024.loc[:,"Dhuhr"] = pd.to_datetime(dfJan2024["Dhuhr"], format="%I:%M %p").dt.time
dfJan2024.loc[:,"Asr"] = pd.to_datetime(dfJan2024["Asr"], format="%I:%M %p").dt.time
dfJan2024.loc[:,"Maghrib"] = pd.to_datetime(dfJan2024["Maghrib"], format="%I:%M %p").dt.time
dfJan2024.loc[:,"Isha"] = pd.to_datetime(dfJan2024["Isha"], format="%I:%M %p").dt.time
```
