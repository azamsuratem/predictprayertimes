```python
import pandas as pd
from io import StringIO
```

```python
## Manual extract HTML Table for Apr 2024
## 1. Browse https://www.google.com/search?q=prayer+times+in+kuala+lumpur
## 2. Tabs "Today" and "30 days", click on "30 days"
## 3. Click "More days" to expand the table, showing all 30 rows
## 4. Inspect element, click tab Console, then type in "document.getElementsByClassName("bvYrdf")[1].innerHTML;" and hit Enter
## 5. Copy the texts appeared below it into the variable GoogleSearchWidgetApr2024 below
##
## MAKE SURE TO RUN THIS ON:
## 2nd of each month to get the first 30 days of the month
## 3rd of the month to get the row data for 31st
```

```python
filePath = 'htmldata\googlesearch_2024_04.txt'

with open(filePath, 'r') as file:
    lines = file.readlines()
    fileContent = ''.join(lines).replace(r'\g+', '')

GoogleSearchWidgetApr2024 = fileContent
```

```python
GoogleSearchDFApr2024 = pd.read_html(StringIO(str(GoogleSearchWidgetApr2024)))[0]
GoogleSearchDFApr2024.iloc[0,0] = "2024"
GoogleSearchDFApr2024.columns = GoogleSearchDFApr2024.iloc[0]
GoogleSearchDFApr2024 = GoogleSearchDFApr2024[1:]
```

```python
GoogleSearchDFApr2024
```

```python
dfApr2024 = GoogleSearchDFApr2024[["2024","Fajr","Dhuhr","Asr","Maghrib","Isha"]]
dfApr2024.loc[:,"2024"] = pd.to_datetime(dfApr2024["2024"].astype(str) + " 2024", format="%d %b %Y").dt.date
dfApr2024.loc[:,"Fajr"] = pd.to_datetime(dfApr2024["Fajr"], format="%H:%M").dt.time
dfApr2024.loc[:,"Dhuhr"] = pd.to_datetime(dfApr2024["Dhuhr"], format="%H:%M").dt.time
dfApr2024.loc[:,"Asr"] = pd.to_datetime(dfApr2024["Asr"], format="%H:%M").dt.time
dfApr2024.loc[:,"Maghrib"] = pd.to_datetime(dfApr2024["Maghrib"], format="%H:%M").dt.time
dfApr2024.loc[:,"Isha"] = pd.to_datetime(dfApr2024["Isha"], format="%H:%M").dt.time
```
